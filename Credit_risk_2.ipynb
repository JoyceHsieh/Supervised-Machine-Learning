```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
train_df
```

```python
test_df
```

# Drop the null columns where all values are null
# Drop the null rows

```python
train_df = train_df.dropna(axis='columns', how='all')
train_df = train_df.dropna()
test_df = test_df.dropna(axis='columns', how='all')
test_df = test_df.dropna()
```

# Convert target column "loan_status" into numeric

```python
risk={"low_risk":0, "high_risk":1}
train_df = train_df.replace({'loan_status': risk})
test_df = test_df.replace({'loan_status': risk})
```

```python
#Check our training data columns
train_df.columns
```

# Create our features
## One-hot-encoding to transform our feature into numeric

```python
# Convert categorical data to numeric and separate target feature for training data
# Create our features
# Convert string columns to numbers and drop 'loan_status' column then assign to X
X_train = train_df.drop(['loan_status'], axis=1)

##One-hot-encoding to transform our feature into numeric
X_train = pd.get_dummies(X_train)


# Create our target
y_train = train_df['loan_status']
X_train.head()
```

```python
y_train.value_counts()
```

```python
test_df.columns
```

```python
# Convert categorical data to numeric and separate target feature for testing data
# Convert categorical data to numeric and separate target feature for training data
# Create our features
# Convert string columns to numbers and drop 'loan_status' column then assign to X
X_test = test_df.drop(['loan_status'], axis=1)

##One-hot-encoding to transform our feature into numeric
X_test = pd.get_dummies(X_test)


# Create our target
y_test = test_df['loan_status']
X_test.head()
```

```python
y_test.value_counts()
```

```python
# add missing dummy variables to testing set
# test_df(2020Q1loans) only have  93 column, use dummies function and add one column into our feature.
dummies=pd.get_dummies(test_df['debt_settlement_flag']).rename(columns={'N':"debt_settlement_flag_Y"})
X_test=pd.concat([X_test, dummies], axis=1)
X_test
```

```python
# Check balances
from collections import Counter
print(Counter(y_train))
print(Counter(y_test))
```

# Create Logistic Regression model

```python
# Train the Logistic Regression model on the unscaled data and print the model score
#Create Logistic Regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
#See the score of our Logistic Regression model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
#Use our Logistic Regression model to test the 2020Q1loans accuracy
predictions = classifier.predict(X_test)
print(f"The new point was classified as: {predictions}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual":y_test})
```

```python
#print the compare reslut of prediction and actual 
predicted_labels = classifier.predict(X_test)
actual_labels = y_test
loan_risk = ['low_risk','high_risk']

def compare_labels(predicted_labels, actual_labels):
    """ a more neatly formatted comparison """
    NUM_LABELS = len(predicted_labels)
    num_correct = 0
    
    for i in range(NUM_LABELS):
        p = int(round(predicted_labels[i]))         # round protects from fp error 
        a = int(round(actual_labels[i]))
        result = "incorrect"
        if p == a:  # if they match,
            result = ""       # no longer incorrect
            num_correct += 1  # and we count a match!

        print(f"row {i:>3d} : {loan_risk[p]:>12s} {loan_risk[a]:<12s}   {result}")   

    print()
    print("Correct:", num_correct, "out of", NUM_LABELS)
    return num_correct

compare_labels(predicted_labels,actual_labels)
```

```python
#Get the score of our Logistic Regression model predict
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
